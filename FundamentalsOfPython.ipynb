```
def Sort(y): 
    y.sort(key = lambda x: x[1]) 
    return y 
  
# Driver Code 
a=[[1,2], [44, 4], [66, 99], [43, 22], [23, -10] ]
print(Sort(a)) 
```

```
#variable length arguments

#keyed argument

def fun(**kwargs):
    print (kwargs["a"], kwargs["b"])
    return

kwargs={"a":5, "b":6}
fun(**kwargs)

#non-keyed arguments

def fun(*arg):
  for a in arg:
    print (a, end=" ")

print ("\n")
fun(5,6,7)
print ("\n")
fun(1,2,3,4)
```

#**Working with Pickle files**

```
import pickle
di = {"Name":"David", "roll_no":1234,"phone":99999999}
li = ["Tom Hanks","Russell Crowe","Leonardo Di Caprio"]
with open('filename.pickle', 'wb') as fp:
    pickle.dump(di, fp)
    pickle.dump(li, fp)

with open('filename.pickle', 'rb') as fp:
    output1 = pickle.load(fp)
    output2 = pickle.load(fp)
print(output1,output2)
```
